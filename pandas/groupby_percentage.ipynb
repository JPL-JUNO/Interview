```python
import pandas as pd
```

描述，计算各个值在其所在组内的占比。

```python
df = pd.DataFrame({'group': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                   'value': [12, 29, 30, 11, 7, 9, 11, 15, 1],})
df
```

```python
df.groupby('group')[['value']].transform(sum)
```

```python
df.assign(percent=df['value'] / df.groupby('group')['value'].transform('sum'))
```
